
##Ejercicio

En el **cuarto paso**, la evaluación de las integrales de convolución para los tres casos se deja abierta como un ejercicio. Tenga en cuenta que, en el primer caso, no se superponen, es decir que $y(t)=0$ para $t<0$.

```python
import numpy as np
t = np.linspace(0, 1, 100)
x_t = np.zeros_like(t)
x_t[(t >= 0) & (t <= 1)] = 1
h_t = np.exp(-t)

# Realizar la convolución
convolution_result = np.convolve(x_t, h_t, mode='full') / len(t)

# Graficar las señales y el resultado de la convolución
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 6))
plt.subplot(311)
plt.plot(t, x_t)
plt.title('Señal x(t)')

plt.subplot(312)
plt.plot(t, h_t)
plt.title('Señal h(t)')

plt.subplot(313)
t_conv = np.linspace(0, 2, len(convolution_result))
plt.plot(t_conv, convolution_result)
plt.title('Resultado de la convolución')

plt.tight_layout()
plt.show()
```
